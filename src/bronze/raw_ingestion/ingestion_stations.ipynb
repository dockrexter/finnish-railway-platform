#### BRONZE INGESTION: STATIONS SNAPSHOT (SCD2 SOURCE)


```python
import requests
import json
import os
from datetime import datetime
```

```python
RAW_ROOT = "/Volumes/rail_lakehouse/bronze_rail/raw_files"
STATIONS_ROOT = f"{RAW_ROOT}/stations"
API_URL = "https://rata.digitraffic.fi/api/v1/metadata/stations"
```

```python
def ingest_stations():
    print("üèõÔ∏è Fetching station metadata...")
    resp = requests.get(API_URL, timeout=15)
    resp.raise_for_status()
    data = resp.json()
    
    snap_date = datetime.now().strftime("%Y-%m-%d")
    folder = f"{STATIONS_ROOT}/date={snap_date}"
    os.makedirs(folder, exist_ok=True)
    
    ts = datetime.now().strftime("%H%M%S")
    file_path = f"{folder}/stations_{ts}.json"
    with open(file_path, "w") as f:
        json.dump(data, f)
    
    print(f"‚úÖ Saved {len(data)} stations to {file_path}")
```

```python
ingest_stations()
```
