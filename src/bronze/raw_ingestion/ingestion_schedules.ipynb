#### BRONZE INGESTION: SCHEDULES (DAILY BATCH)


```python

import requests
import json
import os
from datetime import datetime, timedelta
```

```python
RAW_ROOT = "/Volumes/rail_lakehouse/bronze_rail/raw_files"
SCHED_ROOT = f"{RAW_ROOT}/schedules"
```

```python
def ingest_schedules(target_date: str | None = None):
    if target_date is None:
        # yesterday by default
        target_date = (datetime.now() - timedelta(days=1)).strftime("%Y-%m-%d")
    
    url = f"https://rata.digitraffic.fi/api/v1/schedules?departure_date={target_date}"
    print(f"ðŸ“… Fetching schedules for {target_date}...")
    
    resp = requests.get(url, timeout=20)
    resp.raise_for_status()
    data = resp.json()
    
    folder = f"{SCHED_ROOT}/date={target_date}"
    os.makedirs(folder, exist_ok=True)
    
    ts = datetime.now().strftime("%H%M%S")
    file_path = f"{folder}/schedules_{ts}.json"
    with open(file_path, "w") as f:
        json.dump(data, f)
    
    print(f"âœ… Saved {len(data)} schedules to {file_path}")
```

```python
ingest_schedules()
```
